```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv("data/city_data.csv")
ride = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
cmb_df = pd.merge(city, ride)

# Display the data table for preview
cmb_df
```

```python
city_total=cmb_df['city'].value_counts()
city_total

ride_total=cmb_df['ride_id'].value_counts()
ride_total

cmb_df
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

#Rural x & y coordinates
rur= cmb_df[cmb_df['type'] == 'Rural'].groupby([cmb_df['city']])
x_rur = rur['ride_id'].count()
y_rur = rur['fare'].mean()
d_rur = rur['driver_count'].mean()

#Urban x & y coordinates
sub = cmb_df[cmb_df['type'] == 'Suburban'].groupby([cmb_df['city']])
x_sub = sub['ride_id'].count()
y_sub = sub['fare'].mean()
d_sub = sub['driver_count'].mean()

#Suburban x & y coordinates
urb= cmb_df[cmb_df['type'] == 'Urban'].groupby([cmb_df['city']])
x_urb = urb['ride_id'].count()
y_urb = urb['fare'].mean()
d_urb = urb['driver_count'].mean()

# Build the scatter plots for each city types
plt.figure(figsize=(12,6))

#Rural Scatter PLot
plt.scatter(x_rur, y_rur, label = 'Rural', s=d_rur * 10, color=['yellow'], 
            edgecolor='green', alpha = 0.60, marker='o')
#Urban Scatter PLot
plt.scatter(x_urb, y_urb, label = 'Urban', s=d_urb * 10, color=['lightpink'], 
            edgecolor='green', alpha = 0.60, marker='o')
#Suburban Scatter Plot
plt.scatter(x_sub, y_sub, label = 'Suburban', s=d_sub  * 10, color=['lightblue'], 
            edgecolor='green', alpha = 0.60, marker='o')

# Incorporate the other graph properties
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')

# Create a legend
legend = plt.legend(fontsize = 10, title= 'City Types') 
legend.legendHandles[0]._sizes = [60]
legend.legendHandles[1]._sizes = [60]
legend.legendHandles[2]._sizes = [60]

# Incorporate a text label regarding circle size
plt.text(42,35,'Size of circle relates dirver/city', fontsize = 12)

#Create a plot gridline
plt.grid()

# Save Figure
plt.savefig('plot/PyberRideSharingData_2016.png')

plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_type= cmb_df.groupby(['type'])['fare'].sum()

plt.figure(figsize=(12,6))

# Build Pie Chart
clr = ['yellow', 'lightpink', 'lightblue']
exp = (0,0, 0.3)
plt.pie(fare_type, labels = fare_type.index, autopct = '%2.2f%%', colors = clr, 
        explode = exp, shadow = True, startangle = 190)
plt.axis('equal')
plt.title('% of Total Fares by City Type')


# Save Figure
plt.savefig('plot/Percent_of_Total_Fares_by_City_Type.png')

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
city_type= cmb_df.groupby(['type'])['ride_id'].count()


plt.figure(figsize=(12,6))

# Build Pie Chart
clr = ['yellow', 'lightpink', 'lightblue']
exp = (0,0, 0.3)
plt.pie(city_type, labels = city_type.index, autopct = '%2.2f%%', colors = clr, 
        explode = exp, shadow = True, startangle = 190)
plt.axis('equal')
plt.title('% of Total Riders by City Type')

# Save Figure
plt.savefig('plot/Percent_of_Total_Fares_by_City_Type.png')

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_type= cmb_df.groupby(['city','type'])['driver_count'].mean()
driver_type2= driver_type.groupby(['type']).sum()

plt.figure(figsize=(12,6))

# Build Pie Chart
clr = ['yellow', 'lightpink', 'lightblue']
exp = (0,0, 0.3)
plt.pie(driver_type2, labels = driver_type2.index, autopct = '%2.2f%%', colors = clr, 
        explode = exp, shadow = True, startangle = 190)
plt.axis('equal')
plt.title('% of Total Drivers by City Type')

# Save Figure
plt.savefig('plot/Percent_of_Total_Drivers_by_City_Type.png')

# Show Figure
plt.show()
```
